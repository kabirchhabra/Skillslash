```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

trainingdata = pd.read_csv('AirPassengers.csv')
```

```python
trainingdata.head()
```

```python
plt.rcParams.update({'figure.figsize':(12,6)})
trainingdata['#Passengers'].plot()
```

**Method-1: Differencing and Seasonal differencing**

```python
#differencing meaning y(t) = y(t)-y(t-1)
```

```python
trainingdata['#Passengers_diff'] = trainingdata['#Passengers'] - trainingdata['#Passengers'].shift(1)
```

```python
trainingdata['#Passengers_diff'].dropna().plot()
```

**Method-2: Transformation**

```python
#Calculate log
trainingdata['adj_log'] = np.log(trainingdata['#Passengers'])

#Calculate square_root
trainingdata['adj_sqrt'] = np.sqrt(trainingdata['#Passengers'])

#calculate cubed root
trainingdata['adj_cbrt'] = np.cbrt(trainingdata['#Passengers'])
```

```python
trainingdata['adj_log'].dropna().plot()
```

```python
trainingdata['adj_sqrt'].dropna().plot()
```

```python
trainingdata['adj_cbrt'].dropna().plot()
```

Now combining both differencing and transformation

```python
trainingdata['#Passengers_log_diff'] = trainingdata['adj_log'] - trainingdata['adj_log'].shift(7)
trainingdata['#Passengers_log_diff'].dropna().plot()
```

```python
trainingdata['#Passengers_sqrt_diff'] = trainingdata['adj_sqrt'] - trainingdata['adj_sqrt'].shift(7)
trainingdata['#Passengers_sqrt_diff'].dropna().plot()
```

```python
trainingdata['#Passengers_cbrt_diff'] = trainingdata['adj_cbrt'] - trainingdata['adj_cbrt'].shift(7)
trainingdata['#Passengers_cbrt_diff'].dropna().plot()
```

Also, if we are predicting based on this transformed value, on prediciting we again add the trend value to get actual prediciton.
