LDA: Latent Dirichlet Allocation

+ Topic modelling is a technique to extract the hidden topics from large texts. LDA is one popular algoritm of topic modelling.

+ LDA considers each document as a collection of topics. Once you provide the algorithm with number of topics, all it does is to rearrange the topics distribution and keywords distribution within the topics to obtain good composition of topic-keywords distribution.

+ Topic: A topic is a collection of dominant keywords that are typial representatives, just by looking at the keywords, you can idetnify what the topic is all about.

```python
from nltk import word_tokenize, pos_tag
```

POS Tagging(Parts of Speech)

```python
def nouns(text):
    is_noun = lambda pos: pos[:2] == 'NN'   #pos tag command for nouns 'NN'
    tokenized = word_tokenize(text)      #splits text into tokens using NLTK library
    all_nouns = [word for (word, pos) in pos_tag(tokenized) if is_noun(pos)] 
    return ' '.join(all_nouns)
```

```python
file = open(b'data_clean.pkl', 'rb')
data_clean = pickle.load(file)
data_clean
```

```python
#Apply the nouns function to the transcirpts to filter only on nouns
data_nouns = pd.DataFrame(data_clean.Transcript.apply(nouns))
data_nouns
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = text.ENGLISH_STOP_WORDS.union(['like', 'im', 'know', 'just', 'dont', 'thats', 'right', 'people', 'youre', 'got', 'gonna', 'time', 'think', 'yeah', 'said'])
```

```python
#Recreate a document term matrix with only nouns.
cvn = CountVectorizer(stop_words = stop_words)
data_cvn = cvn.fit_transform(data_nouns.Transcript)
data_dtmn = pd.DataFrame(data_cvn.toarray(), columns= cvn.get_feature_names())
data_dtmn.index = data_nouns.index
data_dtmn
```

```python
#Create gensim corpus
corpusn = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmn.transpose()))

#Create vocabulary dicitonary
id2wordn = dict((v,k) for k,v in cvn.vocabulary_.items())
```

```python
ldan = models.LdaModel(corpus=corpusn, num_topics=2, id2word = id2word,passes=10)
ldan.print_topics()
```

# Topic Modeling
Attempt: 3 (Nouns and Adjectives)

```python
def nouns_adj(text):
    '''Given a string of text, tokenize the text and pull out only the nouns and adjectives.'''
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
data_nouns_adj = pd.DataFrame(data_clean.Transcript.apply(nouns_adj))
data_nouns_adj
```

```python
#Create a new document-term matrix using only nouns and adjectives, also remove common words with max_df
cvna = CountVectorizer(stop_words=stop_words, max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj.Transcript)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names())
data_dtmna.index = data_nouns_adj.index
data_dtmna
```

```python
#create gensim corpus
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

#Create vocabulary dicitonary
id2wordna = dict((v, k) for k, v in cvna.vocabulary_.items())
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=2, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=3, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=10)
ldana.print_topics()
```

```python
corpus_transformed = ldana[corpusna]
list(zip([a for [(a,b)] in corpus_transformed], data_dtmna.index))
```
