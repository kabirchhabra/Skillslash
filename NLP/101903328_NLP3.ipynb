**Frequent 30 words by comedian**

```python
from collections import Counter
```

```python
sp = spacy.load('en_core_web_sm')
all_stopwords = sp.Defaults.stop_words
```

```python
final_list = []
top_words = {}
words = []
for i in range(len(df)):
  split = df['Transcript'][i].split()
  Countersplit = Counter(split)
  most_occur = Countersplit.most_common(30)
  lst = [most_occur[i][0] for i in range(len(most_occur))]
  final_list.append(lst)
  top_words[df.index[i]] = [most_occur[i] for i in range(30) if most_occur[i][0] not in all_stopwords]
```

**Stop words**

```python
import spacy
import nltk
nltk.download('punkt')
```

```python
sp = spacy.load('en_core_web_sm')
```

```python
all_stopwords = sp.Defaults.stop_words
```

```python
final_mean_words = []
final_nmean_words = []
for i in final_join_string:
  text_tokens = nltk.tokenize.word_tokenize(i)
  mean_word = [word for word in text_tokens if not word in all_stopwords]
  nmean_word = [word for word in text_tokens if word in all_stopwords]
  final_mean_words.append(mean_word)
  final_nmean_words.append(nmean_word)  
```

# Assignment-2

**Unique words by each comedian**

```python
import numpy as np
```

```python
from collections import Counter
```

```python
unique_words = []
total_words = []
for i in range(len(df)):
  split = df.Transcript[i].split()
  total_words.append(len(split))
  items = Counter(split).keys()
  unique_words.append(len(items))
```

```python
run_times = [64, 27, 72, 60, 58, 70, 72, 63, 63, 59]
```

```python
unique_df = pd.DataFrame({'Comedians':df.index, 'Unique_words':unique_words, 'Total_words':total_words, 'Run_time':run_times})
```

```python
unique_df['Words/minute'] = unique_df['Total_words']/unique_df['Run_time']
```

```python
unique_df.sort_values('Words/minute', inplace=True)
```

```python
unique_df
```

```python
unique_df.to_pickle('unique_words.pkl')
```

**Plotting the results**

```python
plt.figure(figsize=(16,6))
colrs = ['#0E185F', '#2FA4FF', '#019267', '#F473B9', '#F0A500', '#1B1A17', '#488FB1', '#D82148', '#FFBBBB', '#F76E11']
xindex = np.arange(len(unique_df))
plt.subplot(1,2,1)
plt.barh(xindex, unique_df['Unique_words'], align='center', color=colrs)
plt.yticks(xindex, unique_df['Comedians'])
plt.title('No. of unique words')

plt.subplot(1,2,2)
plt.barh(xindex, unique_df['Words/minute'], align='center', color=colrs)
plt.yticks(xindex, unique_df['Comedians'])
plt.title('No. of words/minute')

plt.tight_layout()
plt.show()
```

# Assignment-3

**Check profanity by analysing the common bad words like fucking, shit, fuck etc.**

```python
bad_words = {'fuck':[], 'shit':[]}
for i in df.Transcript:
  bad_words['fuck'].append(i.count('fuck') + i.count('fucking'))
  bad_words['shit'].append(i.count('shit'))
```

```python
bad_words
```

```python
bad_df = pd.DataFrame(bad_words)
bad_df.set_index(df.index)
```

```python
bad_df.to_pickle('profanity.pkl')
```

```python
plt.figure(figsize=(16,6))
for i in range(len(bad_df)):
  x = bad_df.fuck[i]
  y = bad_df.shit[i]

  plt.scatter(x,y)
  plt.text(x+1.5, y+0.5, df.index[i])
  plt.xlim(-5,155)

plt.title('No. of bad words')
plt.xlabel('f')
plt.ylabel('s')

plt.show()
```
